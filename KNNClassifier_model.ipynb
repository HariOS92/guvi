```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
df=pd.read_csv('trainKNN.txt',header=None)
df_test=pd.read_csv('testKNN.txt',header=None)
df.shape,df_test.shape
df
```

```python
df.iloc[:,-1].value_counts()
df.iloc[:,-2].value_counts()
# df.iloc[:,-3].value_counts()
```

```python
# df.corr()
# df.iloc[:,-2].value_counts()
df.drop(df.columns[:-3],axis=1)
# df.iloc[:,:-3]
df.iloc[:,-3].mean(),df.iloc[:,-2].mean()
```

```python
y_train=df.iloc[:,-1]
# y
df.iloc[:,:-3]
# y_train.shape,x_train.shape
x_train=df.iloc[:,:-3]
# print(df.drop(df.iloc[:,:-1],axis=1))
y_test=df_test.iloc[:,-1]
x_test=df_test.iloc[:,:-3]
x_train
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train=sc.fit_transform(x_test)
sc.fit(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, roc_auc_score
knnEuc = KNeighborsClassifier(n_neighbors=6)
knnEuc.fit(x_train,y_train)
y_pred = knnEuc.predict(x_test)
y_pred_prob=knnEuc.predict_proba(x_test)
print('Accuracy:',accuracy_score(y_test,y_pred))
print('AUROC:',roc_auc_score(y_test,y_pred_prob,multi_class='ovr'))
```

```python
# from sklearn.neighbors import DistanceMetric
# manmet = DistanceMetric.get_metric('manhattan')

knnMan = KNeighborsClassifier(n_neighbors=6,metric='manhattan')
knnMan.fit(x_train,y_train)
ym_pred = knnMan.predict(x_test)
ym_pred_prob=knnEuc.predict_proba(x_test)
print('Accuracy:',accuracy_score(y_test,ym_pred),'\nAUROC:',roc_auc_score(y_test,ym_pred_prob,multi_class='ovr'))
```

With Euclidean and Manhattan metrics, there is 5 percent increase in accuracy of KNN model  but AUROC remains the same.
